```python
# for each media file for each annotator in the test file prepare reference and prediction files
from stutter.utils.annotation import LabelMap
label_map = LabelMap()
df = df[df['split'] == 'test']
df_ann = df[df['annotator'].isin(['A1', 'A2', 'A3'])]

for j in df_ann['annotator'].unique():
    df_annotator = df_ann[df_ann['annotator'] == j]
    if j == 'A3':
        # drop the media_file 24fc
        df_annotator = df_annotator[df_annotator['media_file'] != '24fc']

    ref_paths = []
    pred_paths = []
    for i in df_annotator['media_file'].unique():
    
        df_media = df_annotator[df_ann['media_file'] == i]
        ref_path = f'../datasets/fluencybank/our_annotations/reading/csv/test/{i}_gold.txt'
        pred_path = f'../datasets/fluencybank/our_annotations/reading/csv/test/{i}_{j}.txt'
        ref_paths.append(ref_path)
        pred_paths.append(pred_path)
        df_media['label_list'] = df_media['label'].apply(lambda x: label_map.get_all(x))
        df_media = df_media.explode('label_list')

        # convert start and stop to seconds
        df_media['start'] = df_media['start'] / 1000
        df_media['end'] = df_media['end'] / 1000
        
        df_media['label_list'] = df_media['label_list'].apply(lambda x: label_map.description[x] if x in label_map.description else x)
        # drop the instances where the label is empty string or None
        df_media = df_media[df_media['label_list'].notnull()]
        df_media = df_media[df_media['label_list'] != 'NonVerbal']
        df_media[['start', 'end', 'label_list']].to_csv(f'{pred_path}', sep=',', header=False, index=False)

    with open(f'list_{j}.txt', 'w') as f:
        for ref_path, pred_path in zip(ref_paths, pred_paths):
            f.write(f'{ref_path}\t{pred_path}\n')
        
```

```python
import json
from stutter.utils.annotation import LabelMap
label_map = LabelMap()
import pandas as pd
total_df = pd.read_csv('../datasets/fluencybank/our_annotations/reading/csv/labels.csv')
total_df.drop(columns=['NV'], inplace=True)
for ann in ['A1', 'A2', 'A3']:
    df = total_df[total_df['annotator']==ann]
    df['None_p'] = df.apply(lambda row: 1 if row[label_map.core].sum() == 0 else 0, axis=1)
    df['None_s'] = df.apply(lambda row: 1 if row[label_map.secondary].sum() == 0 else 0, axis=1)
    core_list = label_map.core + ['None_p']
    secondary_list = label_map.secondary + ['None_s']
    tension_list = ['T_0','T_1','T_2','T_3']

    # make one hot encoded vector for the T column
    temp_df = pd.get_dummies(df['T'])
    temp_df.columns = tension_list
    temp_df = temp_df.astype(int)
    df = pd.concat([df, temp_df], axis=1)
    df.head()

    # Group counts
    primary_counts = df[core_list].sum().to_dict()
    secondary_counts = df[secondary_list].sum().to_dict()

    # Generate JSON structure
    json_structure = {
        "primary type": primary_counts,
        "Secondary behaviour": {},
        "Tension":{}
    }

    # Populate secondary behaviour
    for col in core_list:
        if col in df.columns:
            # Extract counts for secondary behaviour
            secondary_values = df.loc[df[col] == 1, secondary_list].sum().to_dict()
            json_structure["Secondary behaviour"][col] = {
                k: secondary_values.get(k, 0) for k in secondary_list if k != col
            }
    # populate tension
    for col in secondary_list:
        if col in df.columns:
            # Extract counts for tension
            tension_values = df.loc[df[col] == 1, tension_list].sum()
            json_structure["Tension"][col] = tension_values.to_dict()

    # Save to JSON file
    with open(f'output_{ann}.json', 'w') as f:
        json.dump(json_structure, f, indent=4)

```

```python
import pandas as pd
from stutter.utils.annotation import LabelMap
label_map = LabelMap()
test_df = pd.read_csv('../datasets/fluencybank/our_annotations/reading/csv/ds5_test_df.csv')
test_df['label'] = test_df.apply(lambda row: label_map.get_all(row[label_map.core+label_map.secondary]+[0]), axis=1)
test_df = test_df.explode('label')
test_df['media_file'] = test_df['file_name'].apply(lambda x: x.split('/')[-1].split('.')[0].split('_')[0])
test_df.dropna(subset=['label'], inplace=True)
```

```python
ref_paths = []
pred_paths = []
for i in test_df['media_file'].unique():
    df_media = test_df[test_df['media_file'] == i]
    pred_path = f'../datasets/fluencybank/our_annotations/reading/csv/test/{i}_ds5.txt'
    ref_path = f'../datasets/fluencybank/our_annotations/reading/csv/test/{i}_gold.txt'
    ref_paths.append(ref_path)
    pred_paths.append(pred_path)
    df_media['label'] = df_media['label'].apply(lambda x: label_map.description[x] if x in label_map.description else x)
    df_media['clip_start'] = df_media['clip_start'] / 1000
    df_media['clip_end'] = df_media['clip_end'] / 1000
    df_media.sort_values(by='clip_start', inplace=True)
    df_media = df_media[df_media['label'].notnull()]
    # remove the nonverbal label
    df_media = df_media[df_media['label'] != 'NonVerbal']
    df_media[['clip_start', 'clip_end', 'label']].to_csv(f'{pred_path}', sep=',', header=False, index=False)


with open(f'list_ds5.txt', 'w') as f:
    for ref_path, pred_path in zip(ref_paths, pred_paths):
        f.write(f'{ref_path}\t{pred_path}\n')
```
